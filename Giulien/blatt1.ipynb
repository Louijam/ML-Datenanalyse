# Aufgabe 1

Quelle: Wetter.de

```python
import matplotlib.pyplot as plt

monate = ["Jan","Feb","Mär","Apr","Mai","Jun","Jul","Aug","Sep","Okt","Nov","Dez"]

temp_de = [2.0,4.0,8.0,13.0,18.0,21.0,23.0,23.0,19.0,13.0,7.0,4.0]
temp_au = [27.0,26.0,23.0,19.0,15.0,11.0,11.0,12.0,15.0,19.0,22.0,25.0]

plt.figure(figsize=(8,4))
plt.plot(monate, temp_de, marker='o', label="Deutschland")
plt.plot(monate, temp_au, marker='s', label="Australien")
plt.title("Mittlere Monatstemperaturen (Deutschland vs. Australien)")
plt.xlabel("Monat")
plt.ylabel("Temperatur (°C)")
plt.grid(True)
plt.show()
```

# Aufgabe 2

## a)

```python
kontinente = ["Asien","Afrika","Nordamerika","Südamerika","Antarktika","Europa","Australien"]
flaechen = [44.4,30.3,24.9,17.8,13.2,10.5,8.5]

plt.figure(figsize=(8,4))
plt.bar(kontinente, flaechen, color='blue')
plt.title("Fläche der Kontinente (Mio km²)")
plt.xlabel("Kontinent")
plt.ylabel("Fläche (Mio km²)")
plt.xticks(rotation=30)
plt.show()
```

## b)

```python
flaechen_summe = sum(flaechen)
wasser = 510 - flaechen_summe

labels = kontinente + ["Wasser"]
anteile = flaechen + [wasser]

plt.figure(figsize=(6,6))
plt.pie(anteile, labels=labels, autopct="%1.1f%%", startangle=90)
plt.title("Prozentuale Flächenverteilung: Kontinente vs. Wasser")
plt.show()
```
