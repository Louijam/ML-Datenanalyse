```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
```

## Aufgabe 1

```python
iris_file = Path('iris.csv')
iris = pd.read_csv(iris_file)

features = ['sepal.length', 'sepal.width', 'petal.length', 'petal.width']
species = iris['variety'].unique()

fig, axes = plt.subplots(2, 2, figsize=(12, 8))
for ax, feature in zip(axes.flat, features):
    for sp in species:
        subset = iris[iris['variety'] == sp]
        ax.hist(subset[feature], bins=15, alpha=0.6, label=sp)
    ax.set_xlabel(f"{feature.replace('.', ' ')} (cm)")
    ax.set_ylabel('HÃ¤ufigkeit')
    ax.legend(title='Art')
fig.suptitle('Histogramme der Iris-Merkmale je Klasse')
plt.tight_layout()
```

## Aufgabe 2
### a)

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
for sp, group in iris.groupby('variety'):
    axes[0].scatter(group['sepal.length'], group['sepal.width'], alpha=0.7, label=sp)
    axes[1].scatter(group['petal.length'], group['petal.width'], alpha=0.7, label=sp)

axes[0].set_xlabel('KelchblattlÃ¤nge (cm)')
axes[0].set_ylabel('Kelchblattbreite (cm)')
axes[0].set_title('Kelchblattmasse')
axes[1].set_xlabel('BlÃ¼tenblattlÃ¤nge (cm)')
axes[1].set_ylabel('BlÃ¼tenblattbreite (cm)')
axes[1].set_title('BlÃ¼tenblattmasse')
axes[0].legend(title='Art')
plt.tight_layout()
```

### b)

```python
corr_petal = iris[['petal.length', 'petal.width']].corr().iloc[0, 1]
corr_sepal = iris[['sepal.length', 'sepal.width']].corr().iloc[0, 1]

print(f"The correlation coefficient between petal length (cm) and petal width (cm) attributes is {corr_petal:.5f}.")
print(f"The correlation coefficient between sepal length (cm) and sepal width (cm) attributes is {corr_sepal:.5f}.")
```

## Aufgabe 3
### a)

```python
pv_file = Path('pvtest.csv')
pv = pd.read_csv(pv_file)

pv.head(10)
```

### b)

```python
selected_cols = ['Dci', 'Dcp', 'Dcu', 'Temp1']
pv_selected = pv[selected_cols]
pv_selected.head(10)
```

### c)

```python
zero_mask = (pv_selected == 0).any(axis=1)
pv_filtered = pv_selected.loc[~zero_mask].copy()

print(f"Anzahl Zeilen vor Filter: {len(pv_selected)}, nach Filter: {len(pv_filtered)}")
pv_filtered.head(10)
```

### d)

```python
corr = pv_filtered.corr()

fig, ax = plt.subplots(figsize=(6, 5))
im = ax.imshow(corr, vmin=-1, vmax=1)
ax.set_xticks(range(len(corr.columns)))
ax.set_xticklabels(corr.columns)
ax.set_yticks(range(len(corr.columns)))
ax.set_yticklabels(corr.columns)

for i in range(len(corr.columns)):
    for j in range(len(corr.columns)):
        ax.text(j, i, f"{corr.iloc[i, j]:.2f}", ha='center', va='center', fontsize=8, color='black')

cbar = fig.colorbar(im, ax=ax)
cbar.set_label('Korrelationskoeffizient')
ax.set_title('Heatmap der Merkmale Dci, Dcp, Dcu, Temp1')
plt.tight_layout()
```
