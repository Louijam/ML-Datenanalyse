```python
import pandas as pd
```

```python
countries_file = 'countries_population.csv'

countries = pd.read_csv(
        countries_file,
        sep=r"\s+(?=\d)",
        engine='python',
        names=['Country', 'Population'],
        thousands=',',
        quotechar="'",
        header=None
    )
countries['Country'] = countries['Country'].str.strip("'")

print(countries.head())
```

## Aufgabe 2

### a)

```python
from pathlib import Path

bundeslaender = pd.read_csv(
    'bundeslaender.txt',
    sep=r"\s+",
    engine='python'
)

scale_columns = ['female', 'male']
bundeslaender[scale_columns] = (bundeslaender[scale_columns] * 1000).astype(int)

bundeslaender['population'] = bundeslaender['female'] + bundeslaender['male']
bundeslaender['density'] = (bundeslaender['population'] / bundeslaender['area']).round(0)

output_path = Path('bundeslaender_density.csv')
bundeslaender.to_csv(output_path, index=False)
print(bundeslaender.head())
```

### b)

```python
female_majority = bundeslaender[bundeslaender['female'] > bundeslaender['male']].copy()
print(f'Anzahl der Bundesländer mit mehr Frauen als Männern: {len(female_majority)}')
print(female_majority[['land', 'female', 'male']])
```

### c)

```python
dense_states = bundeslaender[bundeslaender['density'] > 1000].copy()
print(f'Bundesländer mit einer Bevölkerungsdichte > 1000 Einwohner/km² = {len(dense_states)}')
print(dense_states[['land', 'density']])
```

### a)

```python
start_workdays = pd.Timestamp('2025-12-24')
end_workdays = pd.Timestamp('2026-01-06')
workdays = pd.date_range(start=start_workdays, end=end_workdays, freq='B')

print(f'Arbeitstage zwischen {start_workdays:%d.%m.%Y} und {end_workdays:%d.%m.%Y}: {len(workdays)}')
print('Einzelne Tage:', ', '.join(day.strftime('%d.%m.%Y') for day in workdays))
```

### b)

```python
start_sundays = pd.Timestamp('2025-12-24')
end_sundays = pd.Timestamp('2027-01-06')
month_starts = pd.date_range(start=start_sundays.normalize(), end=end_sundays.normalize(), freq='MS')
sunday_firsts = [day for day in month_starts if start_sundays <= day <= end_sundays and day.weekday() == 6]

print(f'Sonntage, die auf den 1. eines Monats fallen: {len(sunday_firsts)}')
print('Daten:', ', '.join(day.strftime('%d.%m.%Y') for day in sunday_firsts))
```
