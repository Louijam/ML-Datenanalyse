```python
import pandas as pd

daten = {
    "Wien":   ["Österreich", 414.6, 1805681],
    "Berlin": ["Deutschland", 891.85, 3562166],
    "Zürich": ["Schweiz", 87.88, 378884],
}

index = [ "country", "area", "population" ]

df = pd.DataFrame(daten, index=index).T.stack()

print(df)
```

### b)

```python
df_sorted = df.sort_index()
print(df_sorted)
```

### c)

```python
df_swapped_sorted = df.swaplevel().sort_index()
print(df_swapped_sorted)
```

## Aufgabe 2
### a)

```python
daten = {
    'Gewicht': [65, 58, 58, 45, 43, 99, 68, 60],
    'Größe':  [1.79, 1.65, 1.72, 1.54, 1.50, 1.89, 1.76, 1.75],
}
index = ['Henry', 'Sarah', 'Elke', 'Lulu', 'Vera', 'Toni', 'Maria', 'Chris']

df2 = pd.DataFrame(daten, index=index)
#df2.index = [ name.lower() for name in df2.index ]

print(df2)
```

### b)

```python
bmi_values = df2['Gewicht'] / (df2['Größe'] ** 2)

#print(bmi_values)

normalbereich = df2[(bmi_values > 18.5) & (bmi_values < 25)]

print(normalbereich)
```

### c)

```python
lowercase_e = df2[df2.index.str.contains('e')]
print(lowercase_e)
```

### d)

```python
df2['BMI'] = df2['Gewicht'] / (df2['Größe'] ** 2)
print(df2)
```

### e)

```python
df2_bmi_sorted = df2.sort_values('BMI', ascending=False)
print(df2_bmi_sorted)
```

### f)

```python
res = df2[(df2['BMI'] < 20) & (df2.index.str.contains('e'))]
print(res)
```
