```python
import matplotlib.pyplot as plt


monate = ["Jan", "Feb", "Mär", "Apr", "Mai", "Jun", "Jul", "Aug", "Sep", "Okt", "Nov", "Dez"]
tempDE=[1.5, 6.6, 7.5,10,14.9,16.8,18.9,20,15.3,11,5.1,2.8]
tempAUS=[27.0, 26.5, 25.5, 23.5, 20.5, 17.5, 16.5, 18.0, 20.5, 23.5, 25.5, 27.5]

plt.plot(monate, tempDE, label='Deutschland')
plt.plot(monate, tempAUS, label='Australien')
plt.xlabel('Monat')
plt.ylabel('Temperatur')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
kontinente=['Asien', 'Afrika', 'N-Amerika', 'S-Amerika','Antarktika', 'Europa', 'Australien' ]
groessen=[44.4, 30.3, 24.9,17.8,13.2,10.5,8.5]

gesamtflaeche=510
for groesse in groessen:
    gesamtflaeche-=groesse

kontinente.append('Wasser')
groessen.append(gesamtflaeche)
fig,ax =plt.subplots(figsize=(8,8))
ax.pie(groessen,
    autopct='%1.1f%%',
       startangle=90
       )
ax.legend(kontinente, title="Kontinente", loc="center left", bbox_to_anchor=(1, 0.5))


plt.show()
```
