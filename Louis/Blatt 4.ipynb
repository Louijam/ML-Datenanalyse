```python
import pandas as pd
import matplotlib.pyplot as plt

# Iris laden
df = pd.read_csv("Input/iris.csv")

# Merkmale aus der CSV: (spaltenname, beschriftung)
features = [
    ("sepal.length", "sepal length (cm)"),
    ("sepal.width", "sepal width (cm)"),
    ("petal.length", "petal length (cm)"),
    ("petal.width", "petal width (cm)")
]

# Klassen und Farben
classes = ["Setosa", "Versicolor", "Virginica"]
colors = ["blue", "red", "green"]

# Plot
plt.figure(figsize=(12, 10))

for i, (colname, label) in enumerate(features, start=1):
    plt.subplot(2, 2, i)
    for cls, col in zip(classes, colors):
        subset = df[df["variety"] == cls][colname]
        plt.hist(subset, bins=10, label=cls, color=col)
    plt.title(label)
    plt.legend()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Iris laden
df = pd.read_csv("Input/iris.csv")

plt.figure(figsize=(12, 5))

# Plot 1: sepal length vs sepal width
plt.subplot(1, 2, 1)
plt.scatter(df["sepal.length"], df["sepal.width"], marker="+", color="blue")
plt.xlabel("sepal length (cm)")
plt.ylabel("sepal width (cm)")

# Plot 2: petal length vs petal width
plt.subplot(1, 2, 2)
plt.scatter(df["petal.length"], df["petal.width"], marker="o", color="green")
plt.xlabel("petal length (cm)")
plt.ylabel("petal width (cm)")

plt.tight_layout()
plt.show()
```

### b)

```python
import pandas as pd

df = pd.read_csv("Input/iris.csv")

# 1. sepal length vs sepal width
corr_sepal = df["sepal.length"].corr(df["sepal.width"])

# 2. petal length vs petal width
corr_petal = df["petal.length"].corr(df["petal.width"])

print(f"The correlation coefficient between petal length (cm) and petal width (cm) attributes is {corr_petal:.5f}.")
print(f"The correlation coefficient between sepal length (cm) and sepal width (cm) attributes is {corr_sepal:.5f}.")

```
