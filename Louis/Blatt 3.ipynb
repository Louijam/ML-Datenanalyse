```python
import pandas as pd
fname = "Input/bundeslaender.txt"
column_names=["Land","area", "male", "female"]

exchange_rates=pd.read_csv(
    fname,
    sep=r"\s+",
    header=0,
    #index_col=0
)


df = pd.DataFrame(exchange_rates)
df["population"] = df["male"] + df["female"]
df["density"] = (df["population"] / df["area"])*1000
df["density"] = df["density"].round(0)


print(df)
df.to_csv("Output/personen.csv")
```

### b)

```python
weiblich_mehr = df[df["female"] > df["male"]]
print(weiblich_mehr)

print("Anzahl:", len(weiblich_mehr))
```

### c)

```python
dichte_groesser_1000 = df[df["density"] > 1000]
print(dichte_groesser_1000)

print("Anzahl:", len(dichte_groesser_1000))
```

```python
import pandas as pd
from datetime import date

start = date(2025, 12, 24)
end = date(2026, 1, 6)


arbeitstage = pd.bdate_range(start, end)

print(arbeitstage)

print("Es gibt",
      len(arbeitstage),
      "normale Wochentage/reguläre Arbeitstage in den Weihnachtsferien 2025/26.")
```

### b)

```python
import pandas as pd

# Zeitraum
dates = pd.date_range("2025-12-24", "2027-01-06")

# Bedingungen:
is_sunday = dates.weekday == 6        # 6 = Sonntag
is_first = dates.day == 1             # 1. des Monats

# Filter
result = dates[is_sunday & is_first]

print(result)

print("Vom 24.12.2025 bis zum 6.01.2027 gibt es",
      len(result),
      "Sonntage, die auf den 1. des Monats fallen.")
```
